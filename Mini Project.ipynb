```python

#Coverting percentages from object to float
items=[ '%_of_Households_Corrugated_Iron_Sheets','%_of_Households_Tiles', '%_of_Households_Concrete',
       'Asbestos_%_of_Households_Sheets', '%_of_Households_Grass',
       '%_of_Households_Makuti', '%_of_Households_Tin',
       '%_of_Households_Mud/Dung', '%_of_Households_Other']
for item in items:
    df[item]=df[item].str.rstrip('%').astype('float')/100.0
```

```python
#Using Object Oriented method to plot the Pie chart
%matplotlib inline
import matplotlib.pyplot as plt
labels = table2.index
sizes = table2['mean'] 

fig1, ax1 = plt.subplots(figsize=(12,12))

ax1.pie(sizes, labels=labels,autopct='%2.2f%%',shadow=True)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title("DATA SUMMARY", fontsize=15,style='italic')
ax1.legend(loc='best')
#labels = ['{0} - {1:1.2f} %']
#sort_legend=True
#fig1
```

```python
#Distribution of corrugated Iron sheet per county
df1=df.groupby(['County']).mean()

#Plotting
fig, axes = plt.subplots(figsize=(20,8))
plt.bar(df1.index, df1['%_of_Households_Corrugated_Iron_Sheets'], align='center', color='green')
plt.xlabel('Counties',fontsize=15)
plt.ylabel('%_of_Households_Corrugated_Iron_Sheets', fontsize=15)
plt.title('Distribution of corrugated roof per county',fontsize=20,style='italic')
plt.xticks(rotation=90, fontsize=13)
plt.tight_layout()
plt.show()
```

```python
#Distribution of grass roofing per county

#Plotting
fig, ax1 = plt.subplots(figsize=(20,8))
plt.bar(df1.index,df1['%_of_Households_Grass'], align='center', color='blue')
plt.xlabel('Counties', fontsize=15)
plt.ylabel('%_of_Households_Grass', fontsize=15)
plt.title('Distribution of grass roofed houses per county',fontsize=20,style='italic')
plt.xticks(rotation=90, fontsize=13)
plt.tight_layout()
plt.show()
```

```python
#A plot of corruated Ironsheet and Grass houses
a=df.groupby(['County']).mean()
a.reset_index(inplace=True)
#a['c']=a['%_of_Households_Corrugated_Iron_Sheets']+a['%_of_Households_Grass']

#a1=a.sort_values('c', ascending=False)

#a1[['%_of_Households_Corrugated_Iron_Sheets','%_of_Households_Grass']].plot(kind='bar',figsize = (15,9))

ax = a[['%_of_Households_Corrugated_Iron_Sheets', '%_of_Households_Grass']].plot(kind = 'bar', legend = True, figsize = (15,9),stacked=True)
ax.set_xticklabels(a1['County'])
```

```python
#Distribution of corrugated Iron sheet per urban and rural area
df2=df.groupby(['Province','rural/urban']).mean()
df2.reset_index(inplace=True)
#Using the pivot function to reshape the data into two columns of rural and urban.
df2.pivot("Province", "rural/urban", "%_of_Households_Corrugated_Iron_Sheets").plot(kind='bar', figsize=(18,8))
plt.xlabel('Province')
plt.ylabel('%_of_Households_Corrugated_Iron_Sheets')
plt.title('Comparison of rural and urban corrugated roof per province',fontsize=15,style='italic')
plt.show()
```

```python
#Distribution of corrugated Iron sheet per urban and rural area
df2=df.groupby(['Province','rural/urban']).mean()
df2.reset_index(inplace=True)
#Using the pivot function to reshape the data into two columns of rural and urban.
df2.pivot("Province", "rural/urban", '%_of_Households_Grass').plot(kind='bar', figsize=(18,8))
plt.xlabel('Province')
plt.ylabel('%_of_Households_Grass')
plt.title('Comparison of rural and urban corrugated roof per province',fontsize=15,style='italic')
plt.show()
```
